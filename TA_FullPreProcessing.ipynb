```python
#Proses case folding
import re
def casefolding(review):
    review = review.lower()
    review = review.strip(" ")
    review = re.sub(r'[?|$|.|!_:")(-+,]','',review)
    return review
df_gt25['review_cfd'] = df_gt25['review'].apply(casefolding)
df_gt25.head()
```

```python
# #Proses Tokenizing
# def token(review):
#     nstr = review.split(' ')
#     dat = []
#     a = -1
#     for hu in nstr:
#         a = a+1
#     if hu == '':
#         dat.append(a)
#     p=0
#     b=0
#     for q in dat:
#         b = q - p
#         del nstr[b]
#         p = p + 1
#     return nstr
# dataset['review_tokens'] = dataset ['review'].apply(token)
# dataset.head(10)

import string
import re #regex library
import nltk

#import word tokenize & freqDist from NLTK
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
nltk.download("punkt")
#-----tokenizing---------

def remove_special(text):
    text = text.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ").replace('\\',"")
    text = text.encode('ascii', 'replace').decode('ascii')
    text = ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ", text).split())
    return text
df_gt25['review_cfd'] = df_gt25['review_cfd'].apply(remove_special)

#remove number
def remove_number(text):
    return re.sub(r"\d+", "", text)
df_gt25['review_cfd'] = df_gt25['review_cfd'].apply(remove_number)

#remove punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans("","",string.punctuation))
df_gt25['review_cfd'] = df_gt25['review_cfd'].apply(remove_punctuation)

#remove whitespace leading & trailing
def remove_whitespace_LT(text):
    return text.strip()
df_gt25['review_cfd'] = df_gt25['review_cfd'].apply(remove_whitespace_LT)

#remove multiple whitespace into single whitespace
def remove_whitespace_multiple(text):
    return re.sub('\s+', ' ',text)
df_gt25['review_cfd'] = df_gt25['review_cfd'].apply(remove_whitespace_multiple)

#remove single char
def remove_singl_char(text):
    return re.sub(r"\b[a-zA-Z]\b", "", text)
df_gt25['review_cfd'] = df_gt25['review_cfd'].apply(remove_singl_char)

# NLTK word tokenize
def token(text):
    return word_tokenize(text)
df_gt25['review_tokens'] = df_gt25['review_cfd'].apply(token)

df_gt25.head()
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download("stopwords")

list_stopwords = stopwords.words('indonesian')

list_stopwords.extend(["yg", "dg", "dgn", "ny", "d", "klo", "kalo",
                      "biar", "bikin", "bilang", "gak", "ga", "krn", "nya",
                      "nih", "sih", "si", "tau", "tdk", "tuh", "utk", "ya",
                       "jd", "jgn", "sdh", "aja", "n", "t", "nyg", "hehe", "pen",
                       "u", "nan", "loh", "&amp", "yah", "mah", "tp", "iiiiiiiiiiiiiiiiiiiiiiiiiii",
                       "bangtttttttttbdjwjdbejsnwbwjsjndjejendjsiwjdjwiwjdjdjdiejjdejncnfjwjwbdiejwjdnjeieidjdje",
                       "aaaaaassssssaaaaaasaaaaa","abchsjaujwhwhwhwwhhwhbwbqhqj", "yayayayayayayayayyayatata",
                       "yayayayayyaayayyayayayayaa","xd", "lg", "bgt"])


list_stopwords = set(list_stopwords)

def stopword_removal(review):
    return [word for word in review if word not in list_stopwords]
df_gt25['review_str'] = df_gt25['review_tokens'].apply(stopword_removal)
df_gt25
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}
    
for doc in df_gt25['review_str']:
    for term in doc:
        if term not in term_dict:
            term_dict[term]= ' '
print(len(term_dict))
print("-----------------------")

for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    print(term,":", term_dict[term])

print(term_dict)
print("-----------------------")

#apply stemmed term to dataframe
def get_stemmed_term(doc):
    return [term_dict[term] for term in doc]

df_gt25['review_stem'] = df_gt25 ['review_str'].swifter.apply(get_stemmed_term)
df_gt25
```
